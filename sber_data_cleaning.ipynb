```python
def outliers_iqr(data, label, left=1.5, right=1.5):
    x = data[label]
    quantile_1, quantile_3 = x.quantile(0.25), x.quantile(0.75)
    iqr = quantile_3 - quantile_1
    lower_bound = quantile_1 - iqr * left
    upper_bound = quantile_3 + iqr * right
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned_data = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned_data
```

```python
outliers, cleaned = outliers_iqr(sber_data, 'full_sq', 1, 6)
display(outliers.shape[0])
display(cleaned.shape[0])
```

```python
def outliers_z_score(data, label, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[label] + 1)
    else:
        x = data[label]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * left
    upper_bound = mu + 3 * right
    outliners = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliners, cleaned
```

```python
outliers, cleaned = outliers_z_score(sber_data, 'mkad_km', True, 3, 3.5)
display(outliers.shape[0], cleaned.shape[0])
```

```python
low_info_cols = []
for col in sber_data.columns:
    top_freq = sber_data[col].value_counts(normalize=True).max()
    nunique_ratio = sber_data[col].nunique() / sber_data[col].count()
    if top_freq > 0.95:
        low_info_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    if nunique_ratio > 0.95:
        low_info_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```
